```python
class Animal:
    def __init__ (self,name,age):
        self.name = name
        self.age = age
    def eat(self):
        print(f"{self.name} is eating")
    def sleep(self):
        print(f"{self.name} is sleeping..")
    def make_sound(self):
        print(f"{self.name} makes a sound")
```

```python
class Dog(Animal):
    def __init__(self, name, age,breed):
        super().__init__(name, age)
        self.breed = breed
    
    def make_sound1(self):
        print(f"{self.name} is Barking")

    def fetch(self):
        print(f"{self.name} is fetching the ball")
```

```python
class Cat(Animal):
    def __init__(self,name,age,colour):
        super().__init__(name,age)
        self.colour = colour

    def make_sound2(self):
        print(f"{self.name} meows")

    def scratch(self):
        print(f"{self.name} scratches")
```

```python
dog = Dog("tommy",3,"shitzu")
cat = Cat("whisker",3,"brown")
```

```python
dog.eat()
```

```python
dog.make_sound()
```

```python
dog.make_sound1()
```

```python
cat.scratch()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    def start(self):
        print(f"{self.brand} vehicle started")
```

```python
class Car(Vehicle):
    def __init__(self,brand,model):
        super().__init__(brand)
        self.model = model
    def drive(self):
        print(f"{self.brand} {self.model} is driving")
    
        
```

```python
class ElectricCar(Car):
    def __init__(self,brand,model,battery_charge):
        super().__init__(brand,model)
        self.battery_charge = battery_charge
    def charge(self):
        print(f"{self.brand} {self.model} is charging at {self.battery_charge} kwh")
```

```python
ecar = ElectricCar("Tesla","Model 3",75)
```

```python
ecar.start()
```

```python
ecar.drive()
```

```python
ecar.charge()
```

```python
class Father:
    def __init__(self):
        self.father_name = "Mr.sharma"
    
    def gardening(self):
        print(f"{self.father_name} is gardening")

    def show_father_traits(self):
        print("Father's traits: Hardworking,Patience")

class Mother:
    def __init__(self):
        self.mother_name = "Mrs.sharma"
    
    def cooking(self):
        print(f"{self.mother_name} is cooking")
    
    def show_mother_traits(self):
        print("Mother's Traits: Caring, Creative")
```

```python
class Child(Father,Mother):
    def __init__(self,child_name):
        Father.__init__(self)
        Mother.__init__(self)
        self.child_name = child_name

    def show_talents(self):
        print(f"\n{self.child_name}'s inherited talents:")
        self.gardening()
        self.cooking()
        self.show_father_traits()
        self.show_mother_traits()
        
```

```python
child = Child("Rahul")
```

```python
child.show_father_traits()
```

```python
child.show_talents()
```
