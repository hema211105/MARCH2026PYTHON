```python
age = 17
gpa = 3.85
subjects = ['Math', 'Science', 'Python']
is_passing = True

with open('student_info.txt', 'w') as f:
    f.write('Age: ' + str(age) + '\n')
    f.write(f'GPA: {gpa} \n')
    f.write('subjects:\n')
    for subject in subjects:
        f.write(f" - {subject}\n")
```
