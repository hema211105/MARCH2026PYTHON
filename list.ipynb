```python
#sort functions
course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
student_counts = [350, 420, 380, 500, 450]
mixed_case_courses = ["azure DevOps", "Python basics", "Data science"]
```

```python
course_list.sort()
course_list
```

```python
#sort in descending order
student_counts.sort(reverse=True)
student_counts
```

```python
# 4. Sort mixed case courses case-insensitively
mixed_case_courses.sort(key=str.lower)
mixed_case_courses
```

```python
course_enrollments = [("Data Science", 380), ("Python Basics", 500), ("Azure DevOps", 420)]
course_enrollments.sort(key=lambda item: item[1])
course_enrollments
```

```python
course_enrollments.sort()
course_enrollments
```

```python

course_enrollments.sort(key=lambda item: item[0])
course_enrollments
```

```python
#Descending order based on zeroth element
course_enrollments.sort(key=lambda item: item[1], reverse=True)
course_enrollments
```

```python
#Descending order based on first element
course_enrollments.sort(key=lambda item: item[0], reverse=True)
course_enrollments
```

```python
#Sort list based on length of strings
courses = ["SQL", "Python", "Azure DevOps", "Git"]
courses.sort(key=len)
courses
```
