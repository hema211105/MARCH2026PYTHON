```python
class Car:
    def __init__(self,brand,model,color):
        self.brand = brand
        self.model = model
        self.color = color
        self.speed = 0
    def display_inf0(self):
        print(f"Car: {self.brand} {self.model}")
        print(f"color:{self.color}")
        print(f"Current Speed: {self.speed} km/h")
    def accelerate(self, increase):
        self.speed += increase
        print(f"{self.brand} is not going at {self.speed} km/h")
    def brake(self,decrease):
        self.speed -= decrease
        if self.speed < 0:
            self.speed = 0
        print(f"{self.brand} slowed down to {self.speed} km/h")
```

```python
class Movie:
    def __init__(self,title,director,year,rating):
        self.title = title
        self.director = director
        self.year = year
        self.rating = rating
    def display_info(self):
        print(f"{self.title}({self.year}), directed by {self.director},Rating: {self.rating}")
    def update_rating(self, new_rating):
        if 0  <= new_rating <=10:
            self.rating = new_rating
        else:
            print("Rating must be between 0 and 10")
    def is_classic(self):
        return self.year < 2000
```

```python
movie1 = Movie("Inception","Christopher Nolan",2010,8.8)
movie1.display_info()
movie1.update_rating(9.0)
movie1.is_classic()
```

```python
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
    def __str__(self):
        return(f"{self.title} by {self.author} in {self.year}")
```

```python
class Library:
    def __init__(self):
        self.books = []
    def add_book(self,book):
        for b in self.books:
            if b.title == book.title:
                print("Book already Exists")
                return
        self.books.append(book)
        print(f"Added: {book.title}")
    def remove_book(self,title):
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                print(f"Removed: {title}")
                return
        print("Book not found")
    def display_books(self):
        if not self.books:
            print("No books available")
        else:
            for book in self.books:
                print(book)
    def find_book(self,title):
        for book in self.books:
            if book.title == title:
                return book
        return None
```

```python
lib = Library()

b1 = Book("Harry Potter", "J.K. Rowling", 1896)
b2 = Book("The Hobbit", "J.R.R. Tolkien", 2005)

lib.add_book(b1)
lib.add_book(b2)

lib.display_books()

lib.remove_book("Harry Potter")

lib.display_books()
```
