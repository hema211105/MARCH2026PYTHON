```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass
    @abstractmethod
    def dummy_method(self):
        pass
```

```python
class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick or self_start")
    def stop(self):
        print("Bike stops if we click hand brake")
    def dummy_method(self):
        pass
```

```python
class Car(Vehicle):
    def start(self):
        print("Car starts with kick or self_start")
    def stop(self):
        print("Car stops using hand brake")
    def dummy_method(self):
        pass
    def accelerate(self):
        print("Car is accelerating")
```

```python
bike = Bike()
```

```python
bike.start()
```

```python
car = Car()
```

```python
car.stop()
```

```python
car.accelerate()
```

```python
car.start()
```

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
```

```python
class Circle(Shape):
    def __init__(self,r):
        self.r =r
    def area(self):
        return 3.14*self.r*self.r
    
```

```python
class Rectangle(Shape):
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length * self.breadth
        
```

```python
c=Circle(15)
```

```python
c.area()
```

```python
r = Rectangle(5,10)
```

```python
r.area()
```
