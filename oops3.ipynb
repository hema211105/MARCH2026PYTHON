```python
#instance method
class Calculator:
    def __init__(self,brand):
        self.brand = brand
        self.result = 0
    def add(self,a,b):
        self.result=a+b
        return self.result
    def display_result(self):
        print(f"{self.brand} Calculator Result :{self.result}")
```

```python
calc = Calculator("Casio")
```

```python
calc.add(10,20)
```

```python
calc.display_result()
```

```python
class Employee:
    company_name = "Edukron corp"
    employee_count = 0

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
        Employee.employee_count += 1
    @classmethod
    def change_company_name(cls, new_name):
        cls.company_name = new_name
        print(f"Company name changed to: {cls.company_name}")
    @classmethod
    def get_employee(cls):
        return cls.employee_count
    def display_info(self):
        print(f"Employee: {self.name}, Company: {Employee.company_name}")
```

```python
class MathOperations:
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def mul(a,b):
        return a*b
    @staticmethod
    def is_even(number):
        return number%2 == 0
    @staticmethod
    def factorial(n):
        if n == 0 or n == 1:
            return 1
        result =1
        for i in range(2,n+1):
            result *=i
            return result
```

```python
print(MathOperations.add(5, 3))       
print(MathOperations.mul(4, 2))   
print(MathOperations.is_even(6))      
print(MathOperations.factorial(5)) 
```
