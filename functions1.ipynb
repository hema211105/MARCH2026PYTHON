```python
#modifying global variable
count = 0
def increment():
    global count
    count += 1
    print("count: ", count)
```

```python
increment()
increment()
count
```

```python
#lambda Functions
# Syntax = lambda arguments: expression
```

```python
def add(a,b):
    return a+b
```

```python
add_lambda = lambda a, b: a+b
```

```python
print(add(1,2))
add_lambda(2,3)
```

```python
#lambda with single argument
square = lambda x: x**2
square(25)

```

```python
#lambda with multiple arguments
area = lambda length, width:length*width
```

```python
area(10,2)
```

```python
is_even = lambda x: "Even" if x%2==0 else "odd"
```

```python
is_even(15)
```

```python
is_odd = lambda x: "odd" if x%2!=0 else "even"
```

```python
is_odd(2)
```

```python
is_positive = lambda x: "+ve" if x>0 else "-ve"
is_positive(5)
```

```python
is_negative = lambda x: "-ve" if x<0 else "+ve"
is_negative(5)
```

```python
def square(x):
    return x*x
numbers = [1,2,3]
result = list(map(square,numbers))
result
```

```python
result1 = list(map(lambda x: x*2, numbers))
result1
```

```python
#filter()
numbers = [1,2,3]
results = list(filter(lambda x: x%2==0, numbers))
results
```
