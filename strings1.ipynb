```python
#string replace operation
course_slogan = "Learn Data Engineering with Real-Time Projects"
tech_stack = "Azure DevOps and Python and SQL"
course_slogan.replace("Engineering","Science")
```

```python
#replacing with underscore
course_slogan.replace(" ","_")
```

```python
#replacing and with &
tech_stack.replace("and","&")
```

```python
#replacing only 1st occ of and
tech_stack.replace("and","&",1)
```

```python
#removing all spaces
course_slogan.replace(" ","")
```

```python
#replacing a character that doesn't exist
course_slogan.replace("python", "Java")
```

```python
#deleting a word from string
course_slogan.replace(" Data","")
```

```python
#case-sensitive replacement
text =  "Python python pYthon"
text.replace("python","PYTHON")
```

```python
#replace multiple characters(by chaining)
filename = "project report final.docx"
filename.replace(" ","_").replace(".docx",".pdf")
```

```python
#replacing using count=0 (no change)
tech_stack.replace("and","&",0)
```
