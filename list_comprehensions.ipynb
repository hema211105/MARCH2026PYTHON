```python
nums = [1,2,3,4]
result = []
for n in nums:
    if n%2 == 0:
        result.append("even")
    else:
        result.append("odd")
print(result)
```

```python
result1 = [print("even") if n%2==0 else print("odd") for n in nums]
result1
```

```python
result = [n*2 if n%2 == 0 else n*3 for n in nums]
result
```

```python
nums = [3, -1, 5, -2, 20, 30, -10, -25]
result = []
for i in nums:
    if i>0:
        result.append(i)
    else:
        result.append(0)
result
```

```python
result = [i if i>0 else 0 for i in nums]
result
```

```python
list_of_lists = [[1, 2], [3, 4], [5, 6]]
result = []
for sublist in list_of_lists:
    for item in sublist:
        result.append(item)
result
```

```python
result = [item for sublist in list_of_lists for item in sublist]
result
```

```python
result = [item*2 for sublist in list_of_lists for item in sublist]
result
```
