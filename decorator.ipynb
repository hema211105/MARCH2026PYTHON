```python
import time
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()
        print(f"[TIME] took {end - start:.4f} seconds")
        return result
    return wrapper
```

```python
#function without decorator
def process_data_without_decorator():
    total = 0
    for i in range(1000000):
        total += i
    return total
```

```python
process_data_without_decorator()
```

```python
#function with timer decorator
@timer
def process_data_with_decorator():
    total = 0
    for i in range(10000000):
        total += i
    return total
```

```python
process_data_with_decorator()
```

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Something before function execution")
        original_function()
        print("'something after function execution")
    return wrapper_function
```

```python
def say_hello():
    print("Hello")
    
```

```python
say_hello()
```

```python
@decorator_function
def say_hello():
    print("Hello")
```

```python
say_hello()
```

```python
@timer
def another_function():
    time.sleep(3)
    print("task done")
```

```python
another_function()
```

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Function {func.__name__} called with {args}")
        return func(*args, **kwargs)
    return wrapper
```

```python
def mul(a,b):
    return a*b
def add(a,b):
    return a+b
```

```python
mul(5,10)
```

```python
add(5,10)
```

```python
@logger
def mul(a,b):
    return a*b
@logger
def add(a,b):
    return a+b
```

```python
mul(5,10)
```

```python
def decorator1(func):
  def wrapper():
    print("Decorator 1")
    func()
  return wrapper

def decorator2(func):
  def wrapper():
    print("Decorator 2")
    func()
  return wrapper
```

```python
@decorator1
def say_hi():
    print("Hi. How are you?")
```

```python
say_hi()
```

```python
@decorator2
@decorator1
def say_hi():
    print("Hi.How are you?")
```

```python
say_hi()
```
