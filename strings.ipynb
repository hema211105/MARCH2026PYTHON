```python
#length len()
institute = "Edukron"
course_slogan = "Learn Data Science with Real-Time Projects"
tech_stack = "Azure Devops and Python"
```

```python
name_length = len(institute)
print(f"1. Length of '{institute}': {name_length}")
```

```python
slogan_length = len(course_slogan)
print(f"2. Length of slogan: {slogan_length}")
```

```python
empty_len = len("")
print(f"3. Length of empty string: {empty_len}")
```

```python
spaces = "      "
spaces_len = len(spaces)
print(f"4. Length of '{spaces}': {spaces_len}")
```

```python
password = "pass"
min_len = 8
if len(password) < min_len:
    print(f"6. Password '{password}' is too short (min {min_len} chars).")
```

```python
combined = institute + tech_stack
combined_len = len(combined)
print(f"7. Length of combined string: {combined_len}")
```

```python
multi_line = """Line 1
Line 2"""
multi_len = len(multi_line)
print(f"10. Length of multi-line string:\n'''{multi_line}'''\nis: {multi_len}")
```
