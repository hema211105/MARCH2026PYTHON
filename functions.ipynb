```python
def calculate_average(scores):
    total = sum(scores)
    avg = total/len(scores)
    return avg
print(calculate_average([12,23,34]))
print(calculate_average([89,323,54,]))
print(calculate_average([89,34,465]))
print(calculate_average([89,23,45,56,6,34,23,12,]))
```

```python
def calculate_sum(*numbers):
    print(type(numbers))
    total = 0
    for num in numbers:
        total+=num
    return total
```

```python
numbers = (10,20,30,40,50,60)

total = 0
for num in numbers:
  total += num

print(total)
```

```python
#keyword arguments
def student_detail(**details):
  print("Student information: ")
  for key, value in details.items():
    print(f"{key}: {value}")
```

```python
student_detail(name = "Bob", age = 20, grade = "A")
```

```python
## combining all args typs

# order matters: positional, *args, default, **krargs

def complex_function(a,b, *args, c =10, **kwargs):
  print(f"a (position): {a}")
  print(f"b (position): {b}")
  print(f"agrs(variable positional): {args}")
  print(f"c (defaul): {c}")
  print(f"kwargs(vaiable keyword: {kwargs})")
```

```python
complex_function(1,2,3,4,5,c=20, x=100, y=200, z= 300)
```
